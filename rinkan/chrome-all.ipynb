```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
df = pd.read_csv('url.csv')

url_list = []

for idx in range(len(df['URL'])):
  url_list.append(df['URL'][idx])
```

```python
page_url = []
images = []
proucts_name = []
other_price = []
rinkan_price = []

browser = webdriver.Chrome(service=Service("/Users/kii/lang/python/chromedriver"))
for url in url_list:
  browser.get(url)

  if url == 'https://www.gsc-rinkan.com/brand/chromehearts':
    elem_kaitori = browser.find_element(by=By.CLASS_NAME, value='is-figure-square')
  else:
    elem_kaitori = browser.find_element(by=By.CLASS_NAME, value='product-list')

  elems = elem_kaitori.find_elements(by=By.CLASS_NAME, value='is-2-tablet')

  for elem in elems:
    page_url.append(url)

    elem = elem.find_element(by=By.CLASS_NAME, value='product-list-item')
    elem1 = elem.find_elements(by=By.TAG_NAME, value='div')[0]
    elem1 = elem1.find_element(by=By.CLASS_NAME, value='product-list-item-figure')
    elem1 = elem1.find_element(by=By.TAG_NAME, value='img')
    images.append('=IMAGE("' + elem1.get_attribute('src') + '", 4, 60, 60)')

    elem2 = elem.find_elements(by=By.TAG_NAME, value='div')[1]
    prouct_name = elem2.find_element(by=By.CLASS_NAME, value='product-list-item-name')

    if url == 'https://www.gsc-rinkan.com/brand/chromehearts':
      other = elem2.find_element(by=By.CLASS_NAME, value='product-list-item-price-01')
    else: 
      other = elem2.find_element(by=By.CLASS_NAME, value='is-size-6')

    rinkan = elem2.find_element(by=By.CLASS_NAME, value='product-list-item-price-02')
    proucts_name.append(prouct_name.text)
    other =  other.text.split(' ')[1].replace('-', '')
    other_price.append(other)
    rinkan = rinkan.text.split(' ')[1].replace('-', '')
    rinkan_price.append(rinkan)
```

```python
df = pd.DataFrame()
df['URL'] = page_url
df['商品名'] = proucts_name
df['他社金額'] = other_price
df['RINKAN買取金額'] = rinkan_price
df['画像URL'] = images
df.to_csv('RINKANクロムハーツ買取実績一覧.csv', index=False)
```
