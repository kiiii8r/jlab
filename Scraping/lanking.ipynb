```python
elems_rankingBox = browser.find_elements(by=By.CLASS_NAME, value='u_areaListRankingBox')

#全ての観光地名を格納するリスト titlesを用意
titles=[]

for elem_rankingBox in elems_rankingBox:
    elem_title = elem_rankingBox.find_element(by=By.CLASS_NAME, value='u_title')
    title = elem_title.text.split('\n')[1]
    titles.append(title)
```

```python
elems_item = browser.find_elements(by=By.CLASS_NAME, value='u_categoryTipsItem')

enjoyments = []

for elem_item in elems_item:
    elem_items = elem_item.find_elements(by=By.CLASS_NAME, value='is_rank')[0]
    elem_item = elem_items.find_element(by=By.CLASS_NAME, value='evaluateNumber')
    enjoyment = elem_item.text
    enjoyments.append(enjoyment)
    
congestions = []

for elem_item in elems_item:
    elem_items = elem_item.find_elements(by=By.CLASS_NAME, value='is_rank')[1]
    elem_item = elem_items.find_element(by=By.CLASS_NAME, value='evaluateNumber')
    congestion = elem_item.text
    congestions.append(congestion)
    
sceneries = []

for elem_item in elems_item:
    elem_items = elem_item.find_elements(by=By.CLASS_NAME, value='is_rank')[2]
    elem_item = elem_items.find_element(by=By.CLASS_NAME, value='evaluateNumber')
    scenery = elem_item.text
    sceneries.append(scenery)

accessies = []

for elem_item in elems_item:
    elem_items = elem_item.find_elements(by=By.CLASS_NAME, value='is_rank')[3]
    elem_item = elem_items.find_element(by=By.CLASS_NAME, value='evaluateNumber')
    access = elem_item.text
    accessies.append(access)
```

```python
elems_item = browser.find_elements(by=By.CLASS_NAME, value='u_categoryTipsItem')

enjoyments = []
congestions = []
sceneries = []
accessies = []


for elem_item in elems_item:
    items = elem_item.find_elements(by=By.CLASS_NAME, value='is_rank')
    for idx in range(len(items)):
        item = float(items[idx].find_element(by=By.CLASS_NAME, value='evaluateNumber').text) 
        if idx == 0:
            enjoyments.append(item)
        elif idx == 1:
            congestions.append(item)
        elif idx == 2:
            sceneries.append(item)
        else:
            accessies.append(item)
```

```python
elems_rank = browser.find_elements(by=By.CLASS_NAME, value='u_rankBox')

ranks = []

for elem_rank in elems_rank:
    elem_rank = elem_rank.find_element(by=By.CLASS_NAME, value='evaluateNumber')
    rank = float(elem_rank.text)
    ranks.append(rank)
ranks
```

```python
elems = browser.find_elements(by=By.CLASS_NAME, value='u_categoryTipsItem')

categories = []

for elem in elems:
    elems_rank = elem.find_elements(by=By.CLASS_NAME, value='is_rank')
    _ranks = []
    for elem_rank in elems_rank:
        rank = elem_rank.find_element(by=By.CLASS_NAME, value='evaluateNumber').text
        _ranks.append(float(rank))
    categories.append(_ranks)
```

```python
###全ページの情報を取得
```

```python
i=1
'https://scraping-for-beginner.herokuapp.com/ranking/?page={}'.format(i)
```

```python
titles = []
ranks = []
categories = []

for page in range(1,4):
    url = 'https://scraping-for-beginner.herokuapp.com/ranking/?page={}'.format(page)
    browser.get(url)
    
    #観光地名
    elems_rankingBox = browser.find_elements(by=By.CLASS_NAME, value='u_areaListRankingBox')
    for elem_rankingBox in elems_rankingBox:
        elem_title = elem_rankingBox.find_element(by=By.CLASS_NAME, value='u_title')
        title = elem_title.text.split('\n')[1]
        titles.append(title)

    #総合評価
    elems_rank = browser.find_elements(by=By.CLASS_NAME, value='u_rankBox')
    for elem_rank in elems_rank:
        elem_rank = elem_rank.find_element(by=By.CLASS_NAME, value='evaluateNumber')
        rank = float(elem_rank.text)
        ranks.append(rank)

    #各項目
    elems = browser.find_elements(by=By.CLASS_NAME, value='u_categoryTipsItem')
    for elem in elems:
        elems_rank = elem.find_elements(by=By.CLASS_NAME, value='is_rank')
        _ranks = []
        for elem_rank in elems_rank:
            rank = elem_rank.find_element(by=By.CLASS_NAME, value='evaluateNumber').text
            _ranks.append(float(rank))
        categories.append(_ranks)
```

```python
all_df = pd.DataFrame()
```

```python
all_df['観光地名'] = titles
all_df['総合評価'] = ranks
df_cat = pd.DataFrame(categories)
df_cat.columns = ['楽しさ','人混みの多さ','景色','アクセス',]
df = pd.concat([all_df, df_cat], axis=1)
```
