```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome(service=Service("/Users/kii/lang/python/chromedriver"))
browser.get('https://scraping-for-beginner.herokuapp.com/ranking/')
```

```python
browser.find_element(by=By.CLASS_NAME, value='brand-logo').click()
```

```python
xpath = '/html/body/div[2]/div/div/div[1]/ul/li[1]'
browser.find_element(by=By.XPATH, value=xpath).text
```
